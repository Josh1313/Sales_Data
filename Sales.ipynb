```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Exploraremos quien compra mas sin el casado no casado, el data set es binario 0 representa no casados y 1 casados

```python
sns.countplot(x= 'Marital_Status', data=df)
```

Exploraremos en que zona se esta vendiendo mas

```python
sns.countplot(x='Zone', data=df)
```

Ahora exploraremos que clase/categoria de producto se esta vendiendo mas

```python
plt.figure(figsize=(15,10))

plt.xticks(rotation=70)

sns.countplot(x='Product_Category', data=df)
```

Ahora exploraremos que clase de personas estan comprando

```python
plt.figure(figsize=(15,10))

plt.xticks(rotation=70)

sns.countplot(x='Occupation', data=df)
```

Ahora haremos una comparacionde quien compra mas si la mujer o el hombre asi que lo haremso por ocupacion y gender


```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(x='Occupation', hue='Gender' , data=df)
plt.show()
```

Hagamos lo mismo con categoria de productos y genders


```python
plt.figure(figsize=(10,5))
plt.xticks(rotation=90)
sns.countplot(x='Product_Category',hue='Gender' , data=df)
plt.show()
```

Ahora veremos unos venta por estado

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot of sales by state
plt.figure(figsize=(12, 6))
sns.barplot(x='State', y='Amount', data=df)
plt.title('Total Sales Amount by State')
plt.xlabel('State')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()
```

las compras por rangos de edades 

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
sns.countplot(x='Occupation', hue='Age Group' , data=df)
plt.show()
```
